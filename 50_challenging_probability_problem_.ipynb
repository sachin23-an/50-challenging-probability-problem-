### The Problem:
you have a drawer with some red socks and some black socks. When you pick two socks without looking, the chance (probability) that both socks are red is exactly 1/2.

We need to answer two questions:
1.  What's the smallest total number of socks that could be in the drawer?
2.  What's the smallest total number of socks if the number of black socks must be an even number?

### Step 1: Understanding the Probability

When you pick two socks, one after the other, the chance of both being red depends on how many red socks (`r`) and how many total socks (`N`) are in the drawer.

*   The chance of picking the first red sock is `r / N` (number of red socks divided by total socks).
*   After taking out one red sock, there's one less red sock (`r-1`) and one less total sock (`N-1`). So, the chance of picking a second red sock is `(r-1) / (N-1)`.
*   To get the chance of *both* happening, we multiply these: `(r / N) * ((r-1) / (N-1))`.

We are told this total chance is 1/2. So, we have the main equation: `r * (r-1) / (N * (N-1)) = 1/2`.
This can be rearranged to `N * (N-1) = 2 * r * (r-1)`.

### Part a: Smallest number of socks (no extra conditions)

**Goal:** Find the smallest whole numbers for `N` (total socks) and `r` (red socks) that fit our equation.

**How we find it:** The notebook uses a simple computer search:
*   It starts checking possible total sock numbers (`N`) from 3 upwards.
*   For each `N`, it tries different numbers of red socks (`r`) from 0 up to `N-1`.
*   It stops as soon as it finds a pair `(N, r)` that makes the equation `N * (N-1) = 2 * r * (r-1)` true.

**The Answer:** The search finds `N = 4` and `r = 3`.
*   This means the drawer has 4 socks in total.
*   3 of these are red socks.
*   This leaves `N - r = 4 - 3 = 1` black sock.

**Confirmation:** With 3 red socks and 1 black sock, if you pick two socks, the chance of both being red is indeed 1/2. (For example, if you label the socks R1, R2, R3, B, there are 12 ways to pick two socks, and 6 of them are both red pairs: R1-R2, R1-R3, R2-R1, R2-R3, R3-R1, R3-R2).

### Part b: Smallest number of socks (if black socks are even)

**Goal:** Find the smallest `N` and `r` that fit our equation, AND also make sure the number of black socks (`b`) is an even number.

**How we find it:** The notebook uses a similar computer search:
*   It uses the same main equation: `N * (N-1) = 2 * r * (r-1)`.
*   It adds an extra check: `(N - r)` must be an even number (meaning it can be divided by 2 without a remainder).
*   It continues searching through `N` and `r` values until it finds the smallest pair that satisfies *both* conditions.

**The Answer:** The search finds `N = 21` and `r = 15`.
*   This means the drawer has 21 socks in total.
*   15 of these are red socks.
*   This leaves `N - r = 21 - 15 = 6` black socks.

**Confirmation:** The number of black socks, 6, is an even number, so this solution fits the new condition. Also, if you calculate the probability with 15 red and 6 black socks, you'll find it's still 1/2.

### Summary:
*   **Smallest socks overall:** 4 socks (3 red, 1 black)
*   **Smallest socks if black socks are even:** 21 socks (15 red, 6 black)

## Solution Overview

This notebook explores a probability problem involving red and black socks in a drawer. The goal is to find the minimum number of socks required for the probability of drawing two red socks to be 1/2. We will first solve this generally, and then with an additional constraint that the number of black socks must be even. The solution uses both mathematical formulas and a brute-force search approach implemented in Python.

### Part a: How small can the number of socks in the drawer be?

The probability of drawing two red socks ($P(2 \text{ red socks})$) from a drawer containing $r$ red socks and $N$ total socks (where $N \ge 2$ and $r \ge 2$) can be calculated as:

$$P(2 \text{ red socks}) = \frac{r}{N} \times \frac{r-1}{N-1}$$

We are given that this probability is 1/2. So, we set up the equation:

$$\frac{1}{2} = \frac{r(r-1)}{N(N-1)} \implies N(N-1) = 2r(r-1)$$

The `loop()` function in the notebook systematically searches for the smallest integer values of $N$ and $r$ that satisfy this equation. It iterates through possible values of $N$ (starting from 3) and for each $N$, it checks values of $r$ from 0 to $N-1$:

```python
def loop():
    found = False
    N = 3
    while not found:
        N += 1
        for r in range(N):
            if N*(N-1)==2*r*(r-1):
                return N, r
```

Executing `loop()` returns `(4, 3)`. This indicates that the smallest number of socks is 4, consisting of 3 red socks and 1 black sock. This result is then verified by manually listing all permutations of drawing two socks from `['R1', 'R2', 'R3', 'B']`, showing that 6 out of 12 possible ordered pairs result in two red socks, confirming the probability of 1/2.

### Part b: How small if the number of black socks is even?

For this part, we introduce an additional constraint: the number of black socks (`b`) must be an even number. Since the total number of socks is $N$ and the number of red socks is $r$, the number of black socks is $b = N - r$. The constraint translates to $(N - r) \% 2 == 0$.

The `loop2()` function extends the logic from `loop()` by adding this new condition to the search criteria:

```python
def loop2():
    found = False
    N = 3
    while not found:
        N += 1
        for r in range(N):
            b = N-r
            if (N*(N-1)==2*r*(r-1))&(b%2==0):
                return N, r
```

When `loop2()` is executed, it returns `(21, 15)`. This means the smallest number of socks that satisfies both the probability condition and the even number of black socks condition is 21. This configuration consists of 15 red socks and $21 - 15 = 6$ black socks. Since 6 is an even number, this solution meets the criteria. The notebook then confirms this result by simulating the draws, showing that the probability of drawing two red socks from a drawer with 15 red and 6 black socks is indeed 1/2.
